```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
#lineplot
x = pd.Series(np.random.normal(size=40))
x.plot()
```

```python
distributions = pd.DataFrame({'normal': np.random.normal(size=15), 
                       'gamma': np.random.gamma(1, size=15), 
                       'poisson': np.random.poisson(size=15)})
```

```python
distributions
```

```python
csum = distributions.cumsum(0)
```

```python
csum.plot()
```

```python
csum.plot(subplots=True, title="Distributions")
```

```python
#KDE plot
distributions.plot(kind='kde', alpha=0.5)
plt.title("Distribution plot")
plt.xlabel("X Axis")
```

```python
#Demonstrating axis limits, linestyle and subplots
fig, axis = plt.subplots(nrows=4, ncols=3, figsize=(17, 10))

#First row
for i,var in enumerate(['normal','gamma','poisson']):
    distributions[var].cumsum(0).plot(color="rgb"[i], kind='bar', ax=axis[0][i], title='Bar', linestyle=[':','-',"-."][i])
    axis[0][i].set_ylabel(var.title())

#Second row
for i,var in enumerate(['normal','gamma','poisson']):
    distributions[var].cumsum(0).plot(color="rgb"[i], kind='line', ax=axis[1][i], title='Line', linestyle=[':','-',"-."][i])
    axis[1][i].set_ylabel(var.title())
    
#Third row
for i,var in enumerate(['normal','gamma','poisson']):
    distributions[var].cumsum(0).plot(color="rgb"[i], kind='kde', ax=axis[2][i], title='Kde', linestyle=[':','-',"-."][i])
    axis[2][i].set_ylabel(var.title())
    
#Forth row
for i,var in enumerate(['normal','gamma','poisson']):
    distributions[var].cumsum(0).plot(color="rgb"[i], kind='kde', ax=axis[3][i], title='Kde with limits set', linestyle=[':','-',"-."][i])
    axis[3][i].set_ylabel(var.title())
    axis[3][i].set_ylim(0,.5)

plt.tight_layout()
```

```python
#Plot
x = np.arange(-10, 10)
y = np.array([i**2 for i in x])

plt.plot(x, y, 'x', color='orange')
print("X:", x)
print("Y:", y)
```

```python
#Scatter
rng = np.random.RandomState(7)
x = rng.randn(100)
y = rng.randn(100)
sizes = 1000 * rng.rand(100)
colors = rng.rand(100)

plt.scatter(x, y, c=colors, s=sizes, alpha=0.5, cmap='Reds')
plt.colorbar()
```

```python
#Histogram
kwargs = dict(bins=30, alpha=0.5, histtype='stepfilled')
plt.hist(x, **kwargs)
plt.hist(y, **kwargs)
```
